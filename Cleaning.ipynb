```python
def cleaning(prem):
    prem.drop(columns=['club_involved_name'])
    prem.isnull().sum().sort_values(ascending=False)
    prem.dropna()
    prem.drop_duplicates()
    
    return prem
```

```python
cleaning(prem)
```

```python
cols = prem.columns # first 30 columns
colours = ['#000099', '#ffff00'] # specify the colours - yellow is missing. blue is not missing.
sns.heatmap(prem[cols].isnull(), cmap=sns.color_palette(colours))
```

```python
for col in prem.columns:
    pct_missing = np.mean(prem[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```
